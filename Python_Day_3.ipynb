# 'If' Statements

The
if statement executes a block of code if a specified condition is
True .

```python
age = 18
if age >= 18:
   print("You are an adult.")
```

# 'Else' Statements

The
else statement executes a block of code if the if condition is
False .

```python
age = 16
if age >= 18:
  print("You are an adult.")
else:
  print("You are not an adult.")
```

# Else If (Elif) Statements


The
elif statement allows you to check multiple conditions. It stands for "else if" and can be
used when you need to check more than one condition.

```python
age = 16
if age >= 18:
  print("You are an adult.")
elif age >= 13:
  print("You are a teenager.")
else:
  print("You are a child.")
```

# Combining If , Elif , and Else Statements


```python
score = 75
if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
else:
  print("Grade: F")
```

# Nested If Statements

You can also nest
if statements within other if statements to check more complex
conditions.

```python
age = 20
has_permission = True

if age >= 18:
  if has_permission:
     print("You can enter the club.")
  else:
     print("You need permission to enter the club.")
else:
  print("You are not allowed to enter the club.")
```

# 'for' Loop in Python

The
for loop in Python is used to iterate over a sequence (such as a list, tuple, dictionary, set,
or string) or other iterable objects.

# Iterating Over a List

```python
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
  print(fruit)
```

# Iterating Over a String


```python

word = "hello"
for letter in word:
  print(letter)
```

# Using range() Function



```python
for i in range(5):
  print(i)
```

# Iterating Over a Dictionary


```python
student_scores = {'Alice': 90, 'Bob': 85, 'Charlie': 92}
for student, score in student_scores.items():
    print(f"{student}: {score}")
```

# Iterating Over a Set


```python
unique_numbers = {1, 2, 3, 4, 5}
for number in unique_numbers:
  print(number)
```

# Using break Statement


```python
for number in range(10):
  if number == 5:
     break
  print(number)
```

# Using continue Statement


```python
for number in range(10):
    if number == 5:
        continue
    print(number)
```

# 'While' Loops in Python

```python
fruits = ['apple', 'banana', 'cherry']

index = 0
while index < len(fruits):
    print(fruits[index])
    index += 1
```

# Iterating Over a String


```python
word = "hello"

index = 0
while index < len(word):
    print(word[index])
    index += 1
```

# Using range() Function


```python
start = 0
end = 5

while start < end:
    print(start)
    start += 1
```

# Iterating Over a Dictionary

```python
student_scores = {'Alice': 90, 'Bob': 85, 'Charlie': 92}

keys = list(student_scores.keys())
index = 0

while index < len(keys):
    key = keys[index]
    print(f"{key}: {student_scores[key]}")
    index += 1
```

# Iterating Over a Set

```python
numbers = {10, 20, 30, 40}

items = list(numbers)
index = 0

while index < len(items):
    print(items[index])
    index += 1
```

# Using break Statement

```python
counter = 0

while counter < 10:
    if counter == 5:
        break
    print(counter)
    counter += 1
```

# Using continue Statement

```python
counter = 0

while counter < 10:
    counter += 1
    if counter % 2 == 0:
        continue
    print(counter)
```

# Logical Operators in Python


Logical operators are used to combine conditional statements. The most common logical
operators in Python are and ,
or , and not

# 1. and Operator

The
and operator returns True if both conditions are
True . If either condition is False , the result is
False .

```python
age = 20
has_permission = True

if age >= 18 and has_permission:
    print("You can enter the club.")
else:
    print("You cannot enter the club.")
```

# 2. or Operator

The
or operator returns
True if at least one of the conditions is
True . If both conditions are
False , the result is
False .

```python
age = 16
has_permission = True

if age >= 18 or has_permission:
    print("You can enter the club.")
else:
    print("You cannot enter the club.")
```

# 3. not Operator

The
not operator inverts the result of the condition. If the condition is True ,
not makes it False , and if the condition is False , not makes it
True .


```python
age = 16

if not age >= 18:
    print("You are not an adult.")
else:
    print("You are an adult.")
```

# 4. Combining Logical Operators

You can combine multiple logical operators to form more complex conditions.

```python
age = 20
has_permission = False
is_vip = True

if (age >= 18 and has_permission) or is_vip:
    print("You can enter the club.")
else:
    print("You cannot enter the club.")
```

# Comparison Operators in Python

Comparison operators are used to compare two values and return a Boolean result (
True or
False ). These operators are essential for making decisions in your code using conditional
statements.

# 1. Equal to (== )

The
== operator checks if two values are equal.


```python
x = 5
y = 5
print(x == y)
```

# 2. Not equal to (!= )

The
!= operator checks if two values are not equal.


```python
x = 5
y = 3
print(x != y)
```

# 3. Greater than (>)

The > operator checks if the value on the left is greater than the value on the right.


```python
x = 7
y = 5
print(x > y)
```

# 4. Less than (<)

The < operator checks if the value on the left is less than the value on the right.


```python
x = 3
y = 5
print(x < y)
```

# 5. Greater than or equal to ( >=)


```python

x = 5
y = 5
print(x >= y)
```

# 6. Less than or equal to (<= )


```python
x = 5
y = 7
print(x <= y)
```
